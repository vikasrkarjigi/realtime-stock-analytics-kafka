```python
import yfinance as yf
import pandas as pd

# Define the stock tickers and corresponding company names
tickers = {
    'META': 'Meta',
    'AMZN': 'Amazon',
    'AAPL': 'Apple',
    'NFLX': 'Netflix',
    'GOOGL': 'Google'
}

# Create an empty DataFrame with the desired column order
columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Company Name']
result_df = pd.DataFrame(columns=columns)

# Fetch and append data for each ticker
for ticker, company_name in tickers.items():
    stock = yf.Ticker(ticker)
    hist = stock.history(period='1mo')  # Get 1 month of historical data

    # Process each row and add to result_df
    for index, row in hist.iterrows():
        new_row = {
            'Date': index.date(),
            'Open': row['Open'],
            'High': row['High'],
            'Low': row['Low'],
            'Close': row['Close'],
            'Volume': row['Volume'],
            'Company Name': company_name
        }
        result_df = pd.concat([result_df, pd.DataFrame([new_row])], ignore_index=True)

# Display the first few rows (nicely formatted in Jupyter)
result_df.head()
```

```python
import yfinance as yf
import pandas as pd

# Define tickers and company names
tickers = {
    'META': 'Meta',
    'AMZN': 'Amazon',
    'AAPL': 'Apple',
    'NFLX': 'Netflix',
    'GOOGL': 'Google'
}

# List to collect all rows
all_data = []

# Fetch and process stock data
for ticker, company_name in tickers.items():
    stock = yf.Ticker(ticker)
    hist = stock.history(period='1mo')

    for index, row in hist.iterrows():
        all_data.append({
            'Date': index.date(),
            'Open': row['Open'],
            'High': row['High'],
            'Low': row['Low'],
            'Close': row['Close'],
            'Volume': row['Volume'],
            'Company Name': company_name
        })

# Create DataFrame at once — no warnings, faster
result_df = pd.DataFrame(all_data, columns=['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Company Name'])

# Show result
result_df.head()
```
